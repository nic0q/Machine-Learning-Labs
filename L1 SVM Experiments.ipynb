```python
import scipy.io
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import numpy as np
from sklearn.svm import SVC 
from scipy.io import loadmat
from collections import defaultdict 
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform
```

### Lectura de datos
Dado que el dataset dado está en formato matlab (.mat) se lee mediante la biblioteca scipy

```python
base_path = 'emnist-letters.mat'
data = loadmat(base_path)
```

#### Organización y selección de muestras de entrenamiento y testing

```python
X_train_data = data['dataset']['train'][0, 0]['images'][0, 0]
y_train_data = data['dataset']['train'][0, 0]['labels'][0, 0]

X_test_data = data['dataset']['test'][0, 0]['images'][0, 0]
y_test_data = data['dataset']['test'][0, 0]['labels'][0, 0]
```

```python
def organize_by_class(X, y):
    class_dict = {i: [] for i in range(1, 27)}  # Clases de 1 a 26
    for index, label in enumerate(y.ravel()):
        class_dict[label].append(index)
    return class_dict

class_indices_train = organize_by_class(X_train_data, y_train_data)
class_indices_test = organize_by_class(X_test_data, y_test_data)
```

```python
def select_samples(class_dict, num_samples):
    indices = []
    for label, idxs in class_dict.items():
        if len(idxs) < num_samples:
            raise ValueError(f"Clase {label} tiene menos ejemplos ({len(idxs)}) que el número de muestras deseado ({num_samples}).")
        selected_indices = np.random.choice(idxs, num_samples, replace=False)
        indices.extend(selected_indices)
    return indices

# 1000 muestras de entrenamiento y 100 de prueba para cada clase(letra)
train_indices = select_samples(class_indices_train, 1000)
test_indices = select_samples(class_indices_test, 100)
random.seed(42)
```

```python
X_train = X_train_data[train_indices]
y_train = y_train_data[train_indices].ravel()  # Arreglo 1D
X_test = X_test_data[test_indices]
y_test = y_test_data[test_indices].ravel()  # Arreglo 1D
```

## Experimento n°1
Se normalizan los vectores

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Entrenamiento del modelo SVM con kernel rbf

```python
svm_model = SVC(kernel = "rbf")
svm_model.fit(X_train, y_train)
```

### Ajuste de hiperparámetros gamma y C
Por temas de rendimiento se seleccionan los primeros 3500 datos de entrenamiento con su respectiva etiqueta y mediante la función RandomSearchCV se entrena 30 veces usando Cross Validation.

```python
param_distributions = {"gamma": reciprocal(0.001, 0.1), "C": uniform(1, 10)}
rnd_search_cv = RandomizedSearchCV(svm_model, param_distributions, n_iter=10, verbose=2, cv=3)
rnd_search_cv.fit(X_train[:3500], y_train.ravel()[:3500])
```

#### El modelo con mejor desempeño

```python
rnd_search_cv.best_estimator_
```

```python
rnd_search_cv.best_score_
```

Se seleccionan los parámetros C y gamma correspondientes y se entrena nuevamente el modelo con los datos de entrenamiento y prueba.

```python
svm_model_adjusted = SVC(kernel = "rbf", C = 9.357365930430243, gamma = 0.001029618789020042)
svm_model_adjusted.fit(X_train, y_train)
```

#### Accuracy
Función **get_accuracy**, se usa la función predict para obtener las etiquetas predichas por el modelo y usando la matriz de confusión se calcula la precisión de cada categoría

```python
def get_accuracy(X_test, y_test, model):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)     # Matriz de confusión (Resultados reales vs predicciones)
    #print(cm) Imprimir en caso sea necesario (Por temas de espacio no se imprime
    # Calcular la precisión por categoría
    accuracies = cm.diagonal() / cm.sum(axis=1)
    print(classification_report(y_test, y_pred, zero_division=0))
```

#### Accuracy por Categoría y Total de modelo con y sin ajuste de hiperparámetros
Se muestra las accuries por categoría de la letra "A"(1) a la "Z"(26). Y además la accuracy total (Ubicada en la 3era columna)

Accuracy modelo sin ajuste de hiperparámetros

```python
get_accuracy(X_test, y_test, svm_model)
```

Accuracy modelo con ajuste de hiperparámetros

```python
get_accuracy(X_test, y_test, svm_model_adjusted)
```

#### Visualización de un ejemplo con la letra real vs la predicción

```python
def visualize_prediction(X, y, model, index):
    image = X[index].reshape(28, 28)  # Redimensionar el vector a 28x28 para visualización
    image = np.rot90(image, k=-1)
    image = np.fliplr(image)
    true_label_index = y[index]
    predicted_label_index = model.predict([X[index]])[0] 
    true_label = chr(true_label_index + 64)
    predicted_label = chr(predicted_label_index + 64)
    
    plt.imshow(image, cmap='gray')
    plt.title(f'Index: {index}\nTrue Label: {true_label}\nPredicted Label: {predicted_label}')
    plt.colorbar()
    plt.grid(False)
    plt.show()
```

### Gráfico con los experimentos realizados

```python
experimentos = ['Experimento 1', 'Experimento 2', 'Experimento 3']
accuracy_original = [85, 87, 90]  # Precisión antes de ajustar hiperparámetros
accuracy_ajustada = [87, 88, 90]  # Precisión después de ajustar hiperparámetros, el tercer experimento no cambia

x = range(len(experimentos))  # Localización en el eje X para los grupos de barras
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(x, accuracy_original, width, label='Antes de Ajustar')
rects2 = ax.bar([p + width for p in x], accuracy_ajustada, width, label='Después de Ajustar')

ax.set_xlabel('Experimentos')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Comparación de Accuracy antes y después de ajustar hiperparámetros')
ax.set_xticks([p + width / 2 for p in x])
ax.set_xticklabels(experimentos)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # Desplazamiento vertical de texto en puntos
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
